```python
df['votes'] = df['votes'].str.replace(",", "")
```

## 20. Plot the average movie rating over the years (Line Chart)


```python
# Calculate average rating by year
avg_rating_by_year = df.groupby('year')['rating'].mean()

# Create line chart
plt.figure(figsize=(12,6))
plt.plot(avg_rating_by_year.index, avg_rating_by_year.values, marker='o')

# Customize the chart
plt.title('Average Movie Rating Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)

plt.show()
```

# # 21. Plot the number of movies released per year (Line Chart)


```python
# Count movies released each year
movies_per_year = df['year'].value_counts().sort_index()

# Create line chart
plt.figure(figsize=(12,6))
plt.plot(movies_per_year.index, movies_per_year.values, marker='o')

# Customize the chart
plt.title('Number of Movies Released Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Movies')
plt.grid(True)

plt.show()
```

## 22. Plot the distribution of movie ratings (Histogram using Matplotlib)


```python
# Create histogram
plt.figure(figsize=(10,6))
plt.hist(df['rating'], bins=15)

# Customize the chart
plt.title('Distribution of Movie Ratings')
plt.xlabel('Rating')
plt.ylabel('Frequency')

plt.show()
```

## 23. Plot the distribution of movie ratings (Histogram using Seaborn)


```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df['rating'], bins=15, kde=True)

plt.title('Distribution of Movie Ratings')
plt.xlabel('IMDb Rating')
plt.ylabel('Number of Movies')

plt.show()
```

# # 24. Plot the relationship between year and rating (Scatter Plot)


```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.scatterplot(
    data=df,
    x='year',
    y='rating',
    alpha=0.6
)

plt.title('Movie Ratings by Release Year')
plt.xlabel('Year')
plt.ylabel('IMDb Rating')

plt.show()
```

# # 25. Plot the correlation between Rating, Votes, and Gross Income (Heatmap)


```python
# Select relevant columns
corr_data = df[['rating', 'votes', 'gross_income']]

# Correlation matrix
corr_matrix = corr_data.corr()

# Heatmap
plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation: Rating, Votes and Gross Income')
plt.show()
```

# # Q26. Plot the correlation of all numeric columns (Full Heatmap)

```python
numeric_df = df.select_dtypes(include=['number'])

plt.figure(figsize=(12,8))
sns.heatmap(
    numeric_df.corr(),
    annot=True,
    cmap='coolwarm',
    fmt='.2f',
    linewidths=0.5
)

plt.title('Full Correlation Heatmap')
plt.show()
```
