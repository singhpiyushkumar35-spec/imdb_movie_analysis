# Q3. How do we clean the duration column? (e.g., '175 min' → 175)

```python
df["duration"] = df["duration"].str.replace('min', "")
```

```python
df.head()
```

```python
df["duration"] = df["duration"].str.replace(',',"")
```

```python
df.head()
```

```python
df["duration"].astype(int)
```

```python
df.info()
```

# Q4. How do we clean the gross column and votes column?

```python
df['votes'] = df['votes'].str.replace(",", "")
```

```python
df['votes'] = df['votes'].astype(int)
```

```python
df['gross_income'] = df['gross_income'].str.replace(',', "")
```

```python
df['gross_income'] = df["gross_income"].str.replace('$',"")
```

```python
df['gross_income'] = df['gross_income'].str.replace('M', "")
```

```python
df['gross_income'] = df['gross_income'].astype(float)
```

```python
df.head()
```

```python
df.info()
```
